# Part I: Import and Inspect Data

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Rising-Stars-by-Sunshine/stats201-prediction-Haowen/main/data/Queried_Data/Ethereum_value.csv', index_col="Date")
df.head()
```

# Part II: Prepare the Y varible for Regression

## 2.1 Write functions to extract the Y variable for Regression 

```python
Y = df["Open"]
Y.head()
```

# Part III: Create the X variables

## 3.1 Shift the Y to get past values

reference:
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.shift.html

```python
# generate a new variable as the previous 1 observable of your Y variable for regression
df['Open_past'] =df['Open'].shift(1)
df.head()
```

## 3.2 Calculate the Moving Averages

references: 

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rolling.html

https://towardsdatascience.com/moving-averages-in-python-16170e20f6c

```python
df['2-month-ma']=df['Open_past'].rolling(window=2, min_periods=1).mean()
df['6-month-ma'] = df['Open_past'].rolling(window=6, min_periods=1).mean()
df['12-month-ma'] = df['Open_past'].rolling(window=12, min_periods=1).mean()
df.head(15)
```

# Part IV Train and Test Split

*reference*:

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.TimeSeriesSplit.html

```python
from sklearn.model_selection import TimeSeriesSplit
tss = TimeSeriesSplit()
print(tss)
```

```python
# change the train and test split parameters 
tss = TimeSeriesSplit(gap=0, max_train_size=None, n_splits=2, test_size=None)
```

```python
for train_idx, test_idx in tss.split(df):
    print("TRAIN:", train_idx, "TEST:", test_idx)
```

```python
train_idx
```

```python
test_idx
```

```python
train_df = df.iloc[train_idx,:]
test_df =  df.iloc[test_idx,:]
```
